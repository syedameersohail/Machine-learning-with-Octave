### outliers insights

```python
def outliers(data,feature):
    plt.figure(figsize=(8,10))
    sns.boxplot(data[feature])
    plt.show()
```

```python
feature = ['age','creatinine_phosphokinase','ejection_fraction','platelets','serum_creatinine','serum_sodium']
for i in feature:
    outliers(new_df,i)
```

## Outlier handling

```python
def IQR(data,feature):
    IQR=data[feature].quantile(0.75)-data[feature].quantile(0.25)
    e_upper=data[feature].quantile(0.75)+3*(IQR)#extreme outliers(upper)
    e_lower=data[feature].quantile(0.25)-3*(IQR)#extreme outliers(lower)
    n_upper=data[feature].quantile(0.75)+1.5*(IQR)#nominal outliers(upper)
    n_lower=data[feature].quantile(0.25)-1.5*(IQR)#nominal outliers(lower)
    print("Inter Quantile Range         :{}:{}".format(feature,IQR))
    print("Extreme outlier Upper limit  :{}:{}".format(feature,e_upper))
    print("Extreme outlier Lower limit  :{}:{}".format(feature,e_lower))
    print("Nominal outlier Upper limit  :{}:{}".format(feature,n_upper))
    print("Nominal outlier Lower limit  :{}:{}\n".format(feature,n_lower))
```

```python
outlier=["creatinine_phosphokinase","ejection_fraction","platelets","serum_creatinine","serum_sodium"]
for var in outlier:
    IQR(new_df,var)
```

## Outlier handling

```python
new_df['creatinine_phosphokinase']=np.where(new_df['creatinine_phosphokinase']>1280.25,1280.25,new_df['creatinine_phosphokinase'])
new_df['creatinine_phosphokinase']=np.where(new_df['creatinine_phosphokinase']<-581.75,-581.75,new_df['creatinine_phosphokinase'])

new_df['ejection_fraction']=np.where(new_df['ejection_fraction']>67.5,67.5,new_df['ejection_fraction'])
new_df['ejection_fraction']=np.where(new_df['ejection_fraction']<7.5,7.5,new_df['ejection_fraction'])

new_df['platelets']=np.where(new_df['platelets']>440000,440000,new_df['platelets'])
new_df['platelets']=np.where(new_df['platelets']<76000,76000,new_df['platelets'])

new_df['serum_creatinine']=np.where(new_df['serum_creatinine']>2.1499999999999995,2.1499999999999995,new_df['serum_creatinine'])
new_df['serum_creatinine']=np.where(new_df['serum_creatinine']>0.15000000000000024,0.15000000000000024,new_df['serum_creatinine'])

new_df['serum_sodium']=np.where(new_df['serum_sodium']<149,149,new_df['serum_sodium'])
new_df['serum_sodium']=np.where(new_df['serum_sodium']<125,125,new_df['serum_sodium'])
```
